# MIAMRS with non-ensemble mode

```python
'''import some bases we need. Here,AutoML8w is our software base, 
you need to download this software in your experiment environment''' 
import pandas as pd
import numpy as np
import AutoML8w as automl
import sklearn
import time
```

```python
'''The parameters of MIAMRS: 
k: the number of neighbors of the new data(default:10)
N: the number of top models of each neighbor(default:10)
DoEnsembel: ensembel or not(default:True)
data_pre_processing: need data preprocessing or not(default:False)
address_data_feats_featurized: the address of database2 file "data_feats_featurized.csv"
address_pipeline: the address of database3 file "pipelines.json"
address_Top50: the address of database4 file "datasetTop50.csv"
Here: we recommend you set 5<=k<=15,7<=N<=15,DoEnsembel=False for small datasets,because the run time increases with bigger k,N and larger datasets.
We have shown ensembel is only effective for medium or large datasets.'''
MIAMRS = automl.Automl(
    k=5,# the number of neighbors of the new data
    N=10,# the number of top models of each neighbor
    DoEnsembel=False, # ensembel or not
    data_pre_processing=False,# need data preprocessing or not
    address_data_feats_featurized=
    '/root/pmf-automl-master/data_feats_featurized.csv',
    # the address of database2 file "data_feats_featurized.csv"
    address_pipeline='/root/pmf-automl-master/pipelines.json',
    # the address of database3 file "pipelines.json"
    # the address of database4 file "datasetTop50.csv"
    address_Top50='/root/datasetTop50.csv')
```

```python
'''Software testï¼šMIAMRS receives data who is read by pandas but
not numpy'''
## Dataset web:http://archive.ics.uci.edu/ml/datasets/Iris/
iris_data = pd.read_csv('/root/MIAMRS/paper_Experiments/Data Sets/iris.data', header=None)
X = iris_data.iloc[:, :4]
y = iris_data.iloc[:, 4]
X_train, X_test, y_train, y_test = \
        sklearn.model_selection.train_test_split(X, y, random_state=1)
t0 = time.perf_counter()
MIAMRS.fit(X_train, y_train)
y_hat = MIAMRS.predict(X_test)
print('The runtime:', time.perf_counter() - t0)
print("Accuracy score:", sklearn.metrics.accuracy_score(y_test, y_hat))
```

# MIAMRS with ensemble mode

```python
'''import some bases we need. Here,AutoML8w is our software base, 
you need to download this software in your experiment environment''' 
import pandas as pd
import numpy as np
import AutoML8w as automl
import sklearn
import time
```

```python
'''The parameters of MIAMRS: 
k: the number of neighbors of the new data(default:10)
N: the number of top models of each neighbor(default:10)
DoEnsembel: ensembel or not(default:True)
data_pre_processing: need data preprocessing or not(default:False)
address_data_feats_featurized: the address of database2 file "data_feats_featurized.csv"
address_pipeline: the address of database3 file "pipelines.json"
address_Top50: the address of database4 file "datasetTop50.csv"
Here: we recommend you set 3<=k<=8,3<=N<=8,DoEnsembel=True for medium or large datasets,
because the run time increases with bigger k,N and larger datasets.
We have shown ensembel is only effective for medium or large datasets.'''
MIAMRS = automl.Automl(
    k=3,# the number of neighbors of the new data
    N=8,# the number of top models of each neighbor
    DoEnsembel=True, # ensembel or not
    data_pre_processing=False,# need data preprocessing or not
    address_data_feats_featurized=
    '/root/pmf-automl-master/data_feats_featurized.csv',
    # the address of database2 file "data_feats_featurized.csv"
    address_pipeline='/root/pmf-automl-master/pipelines.json',
    # the address of database3 file "pipelines.json"
    # the address of database4 file "datasetTop50.csv"
    address_Top50='/root/datasetTop50.csv')
```

```python
## Dataset web: http://archive.ics.uci.edu/ml/datasets/Gas+sensors+for+home+activity+monitoring#
HT_Sensor_dataset = pd.read_csv('/root/MIAMRS/paper_Experiments/Data Sets/HT_Sensor_dataset.dat',
                                sep='\s+',
                                header=0)
HT_Sensor_matadataset = pd.read_csv('/root/MIAMRS/paper_Experiments/Data Sets/HT_Sensor_metadata.dat',
                                    sep='\s+',
                                    header=0)
labels = list(HT_Sensor_matadataset['class'].unique())
HT_Sensor_matadataset['class'] = HT_Sensor_matadataset['class'].replace(
    labels, list(range(len(labels))))
df = pd.merge(HT_Sensor_dataset,
              HT_Sensor_matadataset[['id', 'class']],
              how='left',
              on='id')
X = df.iloc[:, 1:12]
y = df.iloc[:, 12]
X_train, X_test, y_train, y_test = \
        sklearn.model_selection.train_test_split(X, y, random_state=1)
t0 = time.perf_counter()
MIAMRS.fit(X_train, y_train)
y_hat = MIAMRS.predict(X_test)
print('The runtime:', time.perf_counter() - t0)
print("Accuracy score:", sklearn.metrics.accuracy_score(y_test, y_hat))
```
